Stirling's Approximation for $ln(n!)$ is:

\begin{equation}
ln(n!) \approx nln(n) − n.
\end{equation}

Write a program to output $ln(n!)$ and its Stirling's approximation for $2 < n < 20$. Plot on the same graph the exact value and the approximate value as a function of n. Be sure to lable your curves and axes. **Hint**: the math module (and numpy.math) provides a factorial() method.

**Expected Plot (note this one has 2 < n < 100)**


```python
#Done

import numpy as np
import matplotlib.pyplot as plt

sApproxArray = []
sExactArray = []

def s_Approx():
    for i in range(2, 100):
        s_Approx = i * np.log(i) - i
#         print(s_Approx)
        sApproxArray.append(s_Approx)
def s_Exact():
    for i in range(2, 100):
#         print(np.log(np.math.factorial(i)))
        sExactArray.append(np.log(float(np.math.factorial(i))))


s_Approx()
s_Exact()
for i in range(len(sExactArray)):
    relError = 100*(sApproxArray[i]/sExactArray[i])
    if(relError > 99):
        error = np.array(relError)
        print('n is: ' , i, 'relative error is: ', np.around(error,3))
        break
        

fig = plt.figure()
axes = fig.add_axes([0.15,0.1,0.8,0.8]) #<--- numbers inside array/list reflect dimensions of graph.
axes.plot(sApproxArray, label = "Stirling Approx. ", linestyle = 'dotted')
axes.plot(sExactArray, label = "Exact")
legend = plt.legend(loc = "upper left")
axes.set_xlabel('n')
axes.set_title("Sterling's Approximation vs. Exact");
```

## Problem 6

Using *linspace* Generate 100 equally spaced points from 0 to 1. Then plot the cube of the values generated from *linespace* vs the *linspace* generated values.

**Expected plot(Problem 6)**


```python
import numpy as np
import matplotlib.pyplot as plt

#plot y vs. x where y = x^3.
num = np.linspace(0, 1.0, num = 100)
# print(num)

fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(num, num**3)
ax.set_xlabel('x')
ax.set_ylabel('x^3')
ax.set_title('x^3 vs. x')
    
```
